```python
abs_path = "D:\\Documents\\vivo\\"
kol_file = "VIVO抖音项目-KOL列表（含内容链接等）-0606.xlsx"
```

```python
import sys

sys.path.append("..")
```

```python
import pandas as pd
import json

from utils.timer import timer, Counter
from utils.excels import each_row, save_excel
```

```python
with Counter():
    kol_sheet = pd.read_excel(abs_path + kol_file, "KOL列表")
```

```python
group_keys = []

group_key = ["作者", "抖音ID", "KOL等级", "视频内容", "url",  "video_urls",  "品牌"]
group_keys.append(group_key)
for i in range(1, 6):
    group_keys.append(["%s.%d" % (g_k, i) for g_k in group_key])
    
all_parts = {}
def row_parser(idx, key, row):
    inner_idx = 0
    for g_k in group_keys:
        new_part = {}
        inner_idx += 1
        
        # 切换同一个dict key
        part = row[g_k]
        for i in range(0, len(g_k)):
            new_part[group_key[i]] = part[g_k[i]]
            
        # 处理
        part_key = "%d-%d" % (idx, inner_idx) # new_part["抖音ID"]
        if 'nan' in str(part_key):
            continue
        part_url = new_part["url"]
        if type(new_part["url"]) is not str and type(new_part["video_urls"]) is str:
            if new_part["video_urls"][:4] in '["["':
                new_part["video_urls"] = new_part["video_urls"][2:]
            if new_part["video_urls"].find('"]') < len(new_part["video_urls"]) - 2:
                new_part["video_urls"] = new_part["video_urls"][:new_part["video_urls"].find('"]') + 2]
            part_video_urls = json.loads(new_part["video_urls"].replace("，", ","))

            part_url = part_video_urls[0]
            for t_url in part_video_urls:
                if not t_url.startswith("http://v") and not t_url.startswith("https://api"):
                    part_url = t_url
        new_part["new_url"] = part_url
        
        all_parts[part_key] = new_part
    
    return True
    
each_row(kol_sheet, row_parser, 100)
```

```python
print(len(all_parts))
```

```python
all_df = pd.DataFrame(all_parts)
all_df
```
